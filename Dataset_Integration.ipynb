```python
#Updated label:
# 0: if label=1 and label_hate_speech=0 (hate speech with fake news)
# 1: if label=1 and label_hate_speech=1/2 (non-hate speech with fake news)
# 2: if label=0 and label_hate_speech=0 (hate speech with real news)
# 3: if label=0 and label_hate_speech=1/2 (non-hate speech with real news)
```

```python
for i in range(0, len(df_combined)):
    if df_combined.label[i]==1 and df_combined.label_hate_speech[i]==0:
        df_combined.Updated_label[i]=0
    elif df_combined.label[i]==1 and df_combined.label_hate_speech[i]!=0:
        df_combined.Updated_label[i]=1
    elif df_combined.label[i]==0 and df_combined.label_hate_speech[i]==0:
        df_combined.Updated_label[i]=2
    else:
        df_combined.Updated_label[i]=3
```
